# Actividad uno

```python
dataframe = pd.read_csv('LinReg.csv', header=None) #Extrayendo datos del archivo csv con Pandas
```

```python
x=dataframe[0].values #extrayendo valores de x
y=dataframe[1].values #extrayendo valores de y
```

```python
a1 = (np.dot(x,y) - 100 * np.average(x) * np.average(y)) / (np.sum(x**2) - 100* (np.average(x)**2)) #aplicando operacion
a1 #obteniendo a1
```

```python
a0= np.average(y) - a1 * np.average(x)
a0 #obteniendo a2
```

# Actividad dos

```python
datos= np.genfromtxt('MultiLinReg.csv', delimiter= ',') # Extrayendo datos del archivo csv con Numpy
```

```python
y=datos[:,-1] #Extrayendo todas las filas y la última columna
xwo1=datos[:,:-1] #Extrayendo todas las filas y todas las columnas menos la última
ones=np.ones(y.shape)#Generando un vector fila de unos 
```

```python
x = np.c_[ones,xwo1] #Uniendo vector fila de unos con matriz de x 
```

```python
A = np.dot(np.linalg.inv(np.dot(x.T,x)),np.dot(x.T,y))
A
```

```python
y,np.dot(x,A)
```
